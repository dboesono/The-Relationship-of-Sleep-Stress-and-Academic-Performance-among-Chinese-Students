```python
df_psqi = df.loc[:, '1_psqi':'9_psqi']
df_psqi
```

```python
q9 = []
for i in df_psqi['9_psqi']:
    if i == 'No problem at all':
        q9.append(0)
    elif i == 'Only a very slight problem':
        q9.append(1)
    elif i == 'Somewhat of a problem':
        q9.append(2)
    else:
        q9.append(3)
len(q9)
```

```python
q2 = []
for i in df_psqi['2_psqi']:
    if i <= 15:
        q2.append(0)
    elif 16 <= i <= 30:
        q2.append(1)
    elif 31 <= i <= 60:
        q2.append(2)
    else:
        q2.append(3)
        
q5a = []
for i in df_psqi['5a_psqi']:
    if i == 'Not during the past month':
        q5a.append(0)
    elif i == 'Less than once a week':
        q5a.append(1)
    elif i == 'Once or twice a week':
        q5a.append(2)
    else:
        q5a.append(3)
        
q2_q5a = []
for i in range(len(q2)):
    if q2[i]+q5a[i] == 0:
        q2_q5a.append(0)
    elif 1 <= (q2[i]+q5a[i]) <= 2:
        q2_q5a.append(1)
    elif 3 <= (q2[i]+q5a[i]) <= 4:
        q2_q5a.append(2)
    else:
        q2_q5a.append(3)
len(q2_q5a)
```

```python
q4 = []
for i in df_psqi['4_psqi']:
    if i > 7:
        q4.append(0)
    elif 6 <= i <= 7:
        q4.append(1)
    elif 5 <= i < 6:
        q4.append(2)
    else:
        q4.append(3)
len(q4)
```

```python
q5b = []
for i in df_psqi['5b_psqi']:
    if i == 'Not during the past month':
        q5b.append(0)
    elif i == 'Less than once a week':
        q5b.append(1)
    elif i == 'Once or twice a week':
        q5b.append(2)
    else:
        q5b.append(3)

q5c = []
for i in df_psqi['5c_psqi']:
    if i == 'Not during the past month':
        q5c.append(0)
    elif i == 'Less than once a week':
        q5c.append(1)
    elif i == 'Once or twice a week':
        q5c.append(2)
    else:
        q5c.append(3)

q5d = []
for i in df_psqi['5d_psqi']:
    if i == 'Not during the past month':
        q5d.append(0)
    elif i == 'Less than once a week':
        q5d.append(1)
    elif i == 'Once or twice a week':
        q5d.append(2)
    else:
        q5d.append(3)

q5e = []
for i in df_psqi['5e_psqi']:
    if i == 'Not during the past month':
        q5e.append(0)
    elif i == 'Less than once a week':
        q5e.append(1)
    elif i == 'Once or twice a week':
        q5e.append(2)
    else:
        q5e.append(3)
 
q5f = []
for i in df_psqi['5f_psqi']:
    if i == 'Not during the past month':
        q5f.append(0)
    elif i == 'Less than once a week':
        q5f.append(1)
    elif i == 'Once or twice a week':
        q5f.append(2)
    else:
        q5f.append(3)
        
q5g = []
for i in df_psqi['5g_psqi']:
    if i == 'Not during the past month':
        q5g.append(0)
    elif i == 'Less than once a week':
        q5g.append(1)
    elif i == 'Once or twice a week':
        q5g.append(2)
    else:
        q5g.append(3)
        
q5h = []
for i in df_psqi['5h_psqi']:
    if i == 'Not during the past month':
        q5h.append(0)
    elif i == 'Less than once a week':
        q5h.append(1)
    elif i == 'Once or twice a week':
        q5h.append(2)
    else:
        q5h.append(3)
        
q5i = []
for i in df_psqi['5i_psqi']:
    if i == 'Not during the past month':
        q5i.append(0)
    elif i == 'Less than once a week':
        q5i.append(1)
    elif i == 'Once or twice a week':
        q5i.append(2)
    else:
        q5i.append(3)

output = []
for i in range(len(q5i)):
    output.append(q5b[i] + q5c[i] + q5d[i] + q5e[i] + q5f[i] + q5g[i] + q5h[i] + q5i[i])
    
q5b_q5j = []
for i in output:
    if i == 0:
        q5b_q5j.append(0)
    elif 1 <= i <= 9:
        q5b_q5j.append(1)
    elif 10 <= i <= 18:
        q5b_q5j.append(2)
    else:
        q5b_q5j.append(3)
len(q5b_q5j)
```
